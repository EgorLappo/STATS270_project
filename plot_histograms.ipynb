```python
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def plot_histograms(df, title):
    fig, axs = plt.subplots(3, 2, figsize=(7, 10), constrained_layout=True)

    sns.histplot(df['s'], ax=axs[0, 0], kde=True)
    axs[0,0].set_title('$\\sigma^2$')
    axs[0,0].set_xlabel('$\\sigma^2$')
    axs[0,0].set_ylabel('')
    axs[0,0].axvline(df['s'].mean(), color='red', linestyle='dashed', linewidth=1)

    sns.histplot(df['tau'], ax=axs[0, 1], kde=True)
    axs[0,1].set_title('$\\tau$')
    axs[0,1].set_xlabel('$\\tau$')
    axs[0,1].set_ylabel('')
    axs[0,1].set_xlim(0, 1)
    axs[0,1].axvline(df['tau'].mean(), color='red', linestyle='dashed', linewidth=1)

    sns.histplot(df['mu1'], ax=axs[1, 0], kde=True)
    axs[1,0].set_title('$\\mu_1$')
    axs[1,0].set_xlabel('$\\mu_1$')
    axs[1,0].set_ylabel('')
    axs[1,0].axvline(df['mu1'].mean(), color='red', linestyle='dashed', linewidth=1)

    sns.histplot(df['mu2'], ax=axs[1, 1], kde=True)
    axs[1,1].set_title('$\\mu_2$')
    axs[1,1].set_xlabel('$\\mu_2$')
    axs[1,1].set_ylabel('')
    axs[1,1].axvline(df['mu2'].mean(), color='red', linestyle='dashed', linewidth=1)

    sns.histplot(df['gamma1'], ax=axs[2, 0], kde=True)
    axs[2,0].set_title('$\\gamma_1$')
    axs[2,0].set_xlabel('$\\gamma_1$')
    axs[2,0].set_ylabel('')
    axs[2,0].axvline(df['gamma1'].mean(), color='red', linestyle='dashed', linewidth=1)

    sns.histplot(df['gamma2'], ax=axs[2, 1], kde=True)
    axs[2,1].set_title('$\\gamma_2$')
    axs[2,1].set_xlabel('$\\gamma_2$')
    axs[2,1].set_ylabel('')
    axs[2,1].axvline(df['gamma2'].mean(), color='red', linestyle='dashed', linewidth=1)

    fig.suptitle(f'Histograms of the posterior distributions for {title}', fontsize=16)

    return fig
```

```python
mh = pd.read_csv('mh_samples.csv')
hmc = pd.read_csv('hmc_samples.csv')
gibbs = pd.read_csv('gibbs_samples.csv')
```

```python
plot_histograms(mh, 'MH sampling').savefig('mh_histograms.png');
plot_histograms(hmc, 'HMC sampling').savefig('hmc_histograms.png');
plot_histograms(gibbs, 'Gibbs sampling').savefig('gibbs_histograms.png');
```
